## The number of LEGO releases on a line chart using Matplotlib. <br>
<br>

```python
plt.figure(figsize=(14,8))
plt.title('The number of LEGO releases')
plt.xticks(fontsize=14, rotation=45)
plt.yticks(fontsize=14)
plt.ylabel('Set Number', fontsize=14)
plt.xlabel('Year', fontsize=14)

sets_by_year = sets.groupby('year').count()
sets_by_year['set_num'].head()
plt.plot(sets_by_year.index[:-2], sets_by_year.set_num[:-2])
```

### Let's work out the number of different themes shipped by year. This means we have to count the number of unique theme_ids per calendar year.

```python
themes_by_year = sets.groupby('year').agg({'theme_id': pd.Series.nunique})
```

```python
themes_by_year.rename(columns = {'theme_id':'nr_themes'}, inplace = True) 
themes_by_year.head(10)
```

```python
themes_by_year.tail()
```

## Plot the number of themes released by year and the number of different themes shipped by year on a line Charts with Two Seperate Axes. Only include the full calendar years (i.e., exclude 2020 and 2021).


### Line Charts with Two Seperate Axes

```python
ax1 = plt.gca()
ax2 = ax1.twinx()

# Add styling
ax1.plot(sets_by_year.index[:-2], sets_by_year.set_num[:-2], color='g')
ax2.plot(themes_by_year.index[:-2], themes_by_year.nr_themes[:-2], 'b')

ax1.set_xlabel('Year')
ax1.set_ylabel('Number of Sets', color='green')
ax2.set_ylabel('Number of Themes', color='blue')
```

## How many parts did the average LEGO set released in 1954 compared to say, 2017?

```python
parts_per_set = sets.groupby('year').agg({'num_parts': pd.Series.mean})
parts_per_set.head()
```

```python
parts_per_set.tail()
```

### Scatter Plots in Matplotlib

## Has the size and complexity of LEGO sets increased over time based on the number of parts? Lest's Plot the average number of parts over time using a Matplotlib scatter plot.

```python
plt.figure(figsize=(14,8))
plt.xticks(fontsize=14, rotation=45)
plt.yticks(fontsize=14)
plt.ylabel('Numbers of parts', fontsize=14)
plt.xlabel('Year', fontsize=14)

plt.scatter(parts_per_set.index[:-2], parts_per_set.num_parts[:-2])
```

### Database Schemas, Foreign Keys and Merging DataFrames

The themes.csv file has the actual theme names. The sets .csv has <code>theme_ids</code> which link to the <code>id</code> column in the themes.csv.

```python
themes = pd.read_csv('/content/themes.csv') # has the theme names!
themes.head()
```

```python
themes[themes.name == 'Star Wars']
```

```python
sets[sets.theme_id == 18]
```

```python
sets[sets.theme_id == 209]
```

### Merging (i.e., Combining) DataFrames based on a Key


```python
set_theme_count = pd.DataFrame({'id':set_theme_count.index, 
                                'set_count':set_theme_count.values})
set_theme_count.head()
```

```python
merged_df = pd.merge(set_theme_count, themes, on='id')
merged_df[:3]
```

```python
plt.figure(figsize=(14,8))
plt.xticks(fontsize=14, rotation=45)
plt.yticks(fontsize=14)
plt.ylabel('Nr of Sets', fontsize=14)
plt.xlabel('Theme Name', fontsize=14)

plt.bar(merged_df.name[:10], merged_df.set_count[:10])

```
